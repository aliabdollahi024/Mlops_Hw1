```python
import pandas as pd
import matplotlib.pyplot as plt
from bidi.algorithm import get_display
from arabic_reshaper import reshape
import seaborn as sns
import matplotlib.pyplot as plt
```

# read data


```python
data = pd.read_csv("dataJustZeroOne.csv")
```

```python
import pickle
with open("labels", "rb") as fp:   # Unpickling
  labels = pickle.load(fp)
```

```python
listCount = []
for j in range(10):
    count = 0
    for i in range(len(labels)):
        if labels[i] == j:
            count+=1
    listCount.append(count)

```

```python
cat = ['اجتماعی' , 'استخدام و کاریابی', 'املاک' , 'تجهیزات و صنعتی', 'خانه و آشخانه'  , 'خدمات'  , 'سرگرمی و فراغت','وسایل شخصی'  , 'وسایل نقلیه'  , 'کالای دیجیتال' , ]
```

```python
arabicLabels =[]
for cats in cat:
  arabicLabels.append(get_display(reshape(cats)))  
```

```python
plt.bar(arabicLabels , listCount )
```
