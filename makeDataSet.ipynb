```python
import requests
# from lxml import etree
from bs4 import BeautifulSoup
import csv
```

```python
listHtml = []
listKala = []
count = 0
for link in list(set(listLinks))[10000:20000]:
    try:
        count+=1
        print(count)
        resp = requests.get("https://divar.ir" + link , timeout = 4)
        soup = BeautifulSoup(resp.text)
        listHtml.append(soup)
        # print(link)
        class1 = soup.find_all("div", {"class": "post-page__section--padded"})
        dictFeatures = dict()
        # if len(class1) > 0: 
        #     print(class1[1])
        listRow = soup.find_all("div" , {"class" : "kt-base-row kt-base-row--large kt-unexpandable-row"})
        if len(listRow)>0:
            for index in range(len(listRow)):
                # print(index)
                key = listRow[index].find_all("p"  , {"class" : "kt-base-row__title kt-unexpandable-row__title"})
                value = listRow[index].find_all("p"  , {"class" : "kt-unexpandable-row__value"})
                if len(value) == 0:
                    value = listRow[index].find_all("a"  , {"class" : "kt-unexpandable-row__action kt-text-truncate"})
                if len(value)>0:
                    dictFeatures[key[0].text] = value[0].text 
        listColumn = soup.find_all('div' , {"class" : "kt-group-row-item kt-group-row-item--info-row"})
        # print(len(listColumn))
        if len(listColumn)>0:
            for index in range(len(listColumn)):
                key = listColumn[index].find_all('span' , {"class" : "kt-group-row-item__title kt-body kt-body--sm"})
                value = listColumn[index].find_all('span' , {"class" : "kt-group-row-item__value"})  
                if len(key)>0 and len(value)>0:
                    dictFeatures[key[0].text] = value[0].text 
        listCat = soup.find_all("li", {"class": "kt-breadcrumbs__item"})
        listCategory = []
        for index in range(len(listCat)):
            category = listCat[index].find_all('a' , {'class' : 'kt-breadcrumbs__link'})
            if len(category) == 0:
                category = listCat[index].find_all('span' , {'class' : 'kt-breadcrumbs__link'})
            if len(category) > 0:
                listCategory.append(category[0].text)
        mainTitleList = soup.find_all('div' , {"class" : "kt-page-title__title kt-page-title__title--responsive-sized"})
        # print(mainTitleList)
        secondTitleList = soup.find_all('div' , {"class" : "kt-page-title__subtitle kt-page-title__subtitle--responsive-sized"})
        dictFeatures['category'] = listCategory
        if len(mainTitleList)>0 :
            dictFeatures['mainTitle'] = mainTitleList[0].text
            dictFeatures['secondTitle'] = secondTitleList[0].text
        description = soup.find_all("p" , {"class" : "kt-description-row__text kt-description-row__text--primary"})
        if len(description)>0:
            dictFeatures['description'] = description[0].text
        dictFeatures['link'] = "https://divar.ir" + link
    except:
        print('error')

    

    # print(listCategory)
    # print(dictFeatures)
    listKala.append(dictFeatures)

    
```
